```python
# Scraping the news articles from Ekantipur and storing in csv file.
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
get_url=requests.get("https://ekantipur.com/news")
soup=BeautifulSoup(get_url.text,"html.parser")
```

```python
home_url="https://ekantipur.com/"
```

```python
meta_data=[]
i=0
raw_data=soup.find_all('article',attrs={'class':'normal'})
for i in range(len(raw_data)):
    data=raw_data[i]
    d={}
    d['Headlines']=data.a.text
    d['Description']=data.p.text
    d['Author']=data.find('div',attrs={'class':'author'}).text
    d['URL_link']=home_url+data.a['href']
    meta_data.append(d)
# meta_data
```

```python
import io
```

```python
with io.open("news.csv","w",encoding="utf-8") as f:
    field_names=list(meta_data[0].keys()) 
    write=csv.DictWriter(f,fieldnames=field_names)
    write.writeheader()
    for news in meta_data:
        write.writerow(news)    
```
